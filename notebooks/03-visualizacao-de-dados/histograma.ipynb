## - Criar um gráfico de histograma para visualizar a distribuição da massa dos pinguins.

```python
# Carregando as bibliotecas necessárias para execução da prática.
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
```

### Leitura de dados

```python
df = sns.load_dataset('penguins')
df.head()
```

```python
# Criando o gráfico
ax = sns.histplot(data=df, x='body_mass_g')

# Observação: Esse gráfico trás uma série de problemas que não queremos para nossa visualização.
# Para contornar isso, a seguir será aplicado algumas dicas para melhor visualização.
```

```python
# UTILIZE LINHAS VERTICAIS E HORIZONTAIS
# Para gráficos de histograma, pode ser interessante o uso de linhas verticais e horizontais 
# para facilitar a visualização.
sns.set_style("whitegrid")

# # MELHORE O TITULO E OS LABELS DOS EIXOS
# # Para melhorar a visualização, utilizar labels mais descritivas.
ax = sns.histplot(data=df, x='body_mass_g')
plt.ylabel('Quantidade')
plt.xlabel('Massa (g)')
plt.title('Distribuição da massa dos pinguins')

# # Adicionando uma linha a origem e removando as bordas.
ax.set(frame_on=False)
ax.axhline(0, color="k", clip_on=False)
plt.show()
```
